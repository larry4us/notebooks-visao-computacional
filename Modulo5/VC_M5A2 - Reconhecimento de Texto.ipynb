### Atividade 2 — Bounding boxes com `detail=1`

Com `detail=1`, o `readtext` retorna as coordenadas de cada detecção junto com o texto e a confiança. Usei OpenCV pra desenhar os retângulos direto na imagem.

```python
import os
os.makedirs("../assets/modulo5/images", exist_ok=True)

from PIL import Image, ImageDraw

def criar_imagem(texto, caminho, tamanho=(400, 100)):
    img = Image.new("RGB", tamanho, color="white")
    draw = ImageDraw.Draw(img)
    draw.text((10, 35), texto, fill="black")
    img.save(caminho)

criar_imagem("Residencia em Visao Computacional", "../assets/modulo5/images/teste1.png")
criar_imagem("Python 3.10 | PyTorch", "../assets/modulo5/images/teste2.png")
criar_imagem("1234 5678 9000", "../assets/modulo5/images/teste3.png")

for caminho in ["../assets/modulo5/images/teste1.png", "../assets/modulo5/images/teste2.png", "../assets/modulo5/images/teste3.png"]:
    plt.figure(figsize=(6, 1.5))
    plt.imshow(Image.open(caminho))
    plt.axis("off")
    plt.show()
    resultado = reader.readtext(caminho, detail=0)
    print(f"Resultado: {resultado}\n")
```

```python
import cv2
import numpy as np

def mostrar_com_bbox(caminho_imagem, reader):
    img = np.array(Image.open(caminho_imagem).convert("RGB"))
    resultados = reader.readtext(caminho_imagem, detail=1)  # detail=1 retorna bbox + texto + confiança

    for (bbox, texto, confianca) in resultados:
        pontos = np.array(bbox, dtype=np.int32)
        cv2.polylines(img, [pontos], isClosed=True, color=(0, 200, 0), thickness=2)
        cv2.putText(
            img, f"{texto} ({confianca:.2f})",
            tuple(pontos[0]),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 50, 50), 1
        )

    plt.figure(figsize=(8, 5))
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Bounding boxes: {caminho_imagem}")
    plt.show()

mostrar_com_bbox("../assets/modulo5/images/eng.jpeg", reader)
mostrar_com_bbox("../assets/modulo5/images/num.jpeg", reader)
```
