### Atividade 2 — Variantes do SAM e YOLO para segmentação

O `yolo11n-seg.pt` é o YOLO treinado para segmentação por instância — ele detecta e segmenta objetos individualmente com contornos precisos. O SAM2 é a segunda geração do Segment Anything, com melhor performance e velocidade, especialmente em vídeo.

```python
# YOLO11n-seg — versão da YOLO treinada para segmentação por instância
yolo_seg = ultralytics.YOLO("yolo11n-seg.pt")
results_yolo_seg = yolo_seg("../assets/modulo3/street.jpeg")
results_yolo_seg[0].save(filename="results_yolo_seg.jpg")

# SAM2 — segunda geração do Segment Anything, mais rápida que o SAM original
sam2 = ultralytics.SAM("sam2_b.pt")
results_sam2 = sam2("../assets/modulo3/street.jpeg")
results_sam2[0].save(filename="results_sam2.jpg", boxes=False)

yolo_seg_img = cv2.imread("results_yolo_seg.jpg")
sam2_img = cv2.imread("results_sam2.jpg")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
axes[0].imshow(cv2.cvtColor(yolo_seg_img, cv2.COLOR_BGR2RGB))
axes[0].set_title("YOLO11n-seg — segmentação por instância")
axes[0].axis("off")
axes[1].imshow(cv2.cvtColor(sam2_img, cv2.COLOR_BGR2RGB))
axes[1].set_title("SAM2 — Segment Anything Model 2")
axes[1].axis("off")
plt.tight_layout()
plt.show()
```

### Atividade 3 — Segmentação com torchvision (DeepLabV3)

Sim, o torchvision também tem modelos de segmentação semântica. O DeepLabV3 com backbone ResNet-50 é um clássico para essa tarefa — classifica cada pixel em uma das 21 categorias do Pascal VOC (pessoas, carros, animais etc.). Diferente do SAM, ele não segmenta instâncias individuais, mas rotula regiões da imagem por categoria.

```python
import torchvision.models.segmentation as seg_models
from PIL import Image as PILImage

# DeepLabV3 com backbone ResNet-50 — classifica cada pixel em uma das 21 classes Pascal VOC
deeplabv3 = seg_models.deeplabv3_resnet50(weights=seg_models.DeepLabV3_ResNet50_Weights.DEFAULT)
deeplabv3.eval()

street_pil = PILImage.open("../assets/modulo3/street.jpeg")

preprocess_seg = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
input_tensor = preprocess_seg(street_pil).unsqueeze(0)

with torch.no_grad():
    output = deeplabv3(input_tensor)["out"]

seg_map = output.argmax(dim=1).squeeze().numpy()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].imshow(street_pil)
axes[0].set_title("Imagem original")
axes[0].axis("off")
axes[1].imshow(seg_map, cmap="tab20")
axes[1].set_title("Segmentação semântica — DeepLabV3 (torchvision)")
axes[1].axis("off")
plt.tight_layout()
plt.show()
```
